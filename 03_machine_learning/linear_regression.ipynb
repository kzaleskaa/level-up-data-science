```python
import copy
import numpy as np
import pandas as pd
from sklearn.impute import SimpleImputer
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("../data/03/housing/housing_train.csv")
df
```

```python
df.info()
```

```python
df["NOX"] = df.NOX.isnull().astype("int")
```

```python
df["LSTAT"] = df.LSTAT.isnull().astype("int")
```

```python
df
```

```python
X_train = df.drop(columns=['MEDV'])
y_train = df['MEDV']
```

```python
X_train
```

```python
imputer = SimpleImputer(strategy="median")
```

```python
X_train = imputer.fit_transform(X_train)    
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_train)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
y_pred = model.predict(X_train)

print("Model on all variables without NA")
print(f"Model MSE: {mean_squared_error(y_train, y_pred)}")
print(f"Model RMSE: {np.sqrt(mean_squared_error(y_train, y_pred))}")
print(f"Model MAE: {mean_absolute_error(y_train, y_pred)}")
```
